# Introduction

We will now find objects within the image and calibrate their positions and their magnitudes. We will do this using two useful astronomical tools: astrometry.net and Sextractor. 

Astrometry.net is a web-based tool that will take any image of the sky (even from a cell phone camera) and determine the World Coordinate System (WCS). You do not need any special software to use Astrometry.net, but it helps if you have an account on their server (but not required). You will need access to the internet. Large images make take some time to upload. After the tool runs, you will be given a new image with the WCS embedded in the FITS header. You will also have access to a file containing the positions of the stars used in the calibration as well as a FITs file that contains ONLY the WCS information.

Sextraxtor is a well-known and often used tool to find and measure the characteristics of objects within astronomical images. This is a software package that you will need to download.

## 3. Let's run SExtractor

The SExtractor User's manual can be downloaded from the following link https://www.astromatic.net/pubsvn/software/sextractor/trunk/doc/sextractor.pdf

There are 3 main types of files you need to use for SExtractor:
    
    * default.sex: The configuration of the source measurement code (gotten from the output of "sex -d")
    * default.param: The parameters you want to measure.  Turn off by prefixing a line with comments ("#") ("sex -dp")
    * default.conv: The filter you convolve the image with

You can run sex to dump default files for the first two.

```python
cmd='sex -d'
print subprocess.check_output(cmd,shell=True)
```

```python
cmd='sex -dp'
print subprocess.check_output(cmd,shell=True)
```

Now, make a 2-dimensional kernal to convolve over your image. Signal processing theory shows that the highest point-source signal-to-noise one could achieve is when the pixel data (noisy and raw) is first convolved with a kernal that represents the point-spread function of the data. In this case, we are assuming that the PSF is symmetric with a size that is best represented by 2 image pixels.

Put the following lines into a file called root + default.conv

```python
file = open(root+"default.conv", 'w')
file.write("CONV NORM\n")
file.write("# 3x3 ``all-ground'' convolution mask with FWHM = 2 pixels.\n")
file.write("1 2 1\n")
file.write("2 4 2\n")
file.write("1 2 1\n")
file.close()
```

```python
cmd='cat '+root+'default.conv'
print subprocess.check_output(cmd,shell=True)
```

```python
file = open(root+"default.param", 'w')
file.write("NUMBER                 Running object number                                      \n")
file.write("X_IMAGE                Object position along x                                    [pixel]\n")
file.write("Y_IMAGE                Object position along y                                    [pixel]\n")
file.write("ALPHA_J2000            Right ascension of barycenter (J2000)                      [deg]\n")
file.write("DELTA_J2000            Declination of barycenter (J2000)                          [deg]\n")
file.write("FLUX_APER(1)           # Flux vector within fixed circular aperture(s)            [count]\n")
file.write("FLUXERR_APER(1)        # RMS error vector for aperture flux(es)                   [count]\n")
file.write("FLUX_AUTO              Flux within a Kron-like elliptical aperture                [count]\n")
file.write("FLUXERR_AUTO           RMS error for AUTO flux                                    [count]\n")
file.write("MAG_AUTO               Kron-like elliptical aperture magnitude                    [mag]\n")
file.write("MAGERR_AUTO            RMS error for AUTO magnitude                               [mag]\n")
file.write("KRON_RADIUS            Kron apertures in units of A or B                          \n")
file.write("A_IMAGE                Profile RMS along major axis                               [pixel]\n")
file.write("B_IMAGE                Profile RMS along minor axis                               [pixel]\n")
file.write("THETA_IMAGE            Position angle (CCW/x)                                     [deg]\n")
file.write("ELLIPTICITY            1 - B_IMAGE/A_IMAGE                                        \n")
file.write("FLUX_MAX               Peak flux above background                                 [count]\n")
file.write("FLUX_RADIUS(1)         # Fraction-of-light radii                                  [pixel]\n")
file.write("VIGNET(25,25)          # Pixel data around detection                              [count]\n")
file.write("FLAGS                  Extraction flags                                           \n")
file.close()
```

```python
cmd='cat '+root+'default.param'
print subprocess.check_output(cmd,shell=True)
```

```python
cmd='sex -d > '+root+'default.sex'
subprocess.call(cmd,shell=True)
cmd='cat '+root+'default.sex'
print subprocess.check_output(cmd,shell=True)
```

Now we define several variables that will be used as inputer parameters for SExtractor

```python
sex_im_file=root+'science_reduced_lab_wcs.fits'
sex_conf_file=root+'default.sex'
sex_param_file=root+'default.param'
sex_conv_file=root+'default.conv'
sex_zp='30.'
sex_satur_level='36000'
sex_mem_pixstack='1000000'
sex_check_type='BACKGROUND,FILTERED'
sex_check_name=root+'science_reduced_lab_wcs.CHECK_BACKGROUND.fits'+','+root+'science_reduced_lab_wcs.CHECK_FILTERED.fits'
```

```python
sex_cat_type='FITS_LDAC'
sex_cat_name=root+'science_reduced_lab_cat.ldac'
cmd='sex -c '+sex_conf_file+' -PARAMETERS_NAME '+sex_param_file+' -FILTER_NAME '+sex_conv_file+' -CATALOG_TYPE '+sex_cat_type+' -MAG_ZEROPOINT '+sex_zp+' -SATUR_LEVEL '+sex_satur_level+' -MEMORY_PIXSTACK '+sex_mem_pixstack+' -CHECKIMAGE_TYPE '+sex_check_type+' -CHECKIMAGE_NAME '+sex_check_name+' -CATALOG_NAME '+sex_cat_name+' '+sex_im_file
cmd_args=shlex.split(cmd)
print cmd
proc = subprocess.Popen(cmd_args,stdout=subprocess.PIPE, stderr=subprocess.PIPE)
stdout, stderr = proc.communicate()
print stderr
```

```python
check_data = fits.getdata(root + 'science_reduced_lab_wcs.CHECK_BACKGROUND.fits')
matplotlib.rcParams['figure.figsize'] = (20.0, 16.0)
fig, ax = plt.subplots()
im = ax.imshow(np.log(check_data),cmap=plt.get_cmap('hot'), interpolation='nearest', origin='lower',
               vmin=np.log(1900), vmax=np.log(3000))
fig.colorbar(im)
plt.show()
matplotlib.rcParams.update(inline_rc)
```

```python
sex_cat_type='ASCII_HEAD'
sex_cat_name=root+'science_reduced_lab_cat.dat'
cmd='sex -c '+sex_conf_file+' -PARAMETERS_NAME '+sex_param_file+' -FILTER_NAME '+sex_conv_file+' -CATALOG_TYPE '+sex_cat_type+' -MAG_ZEROPOINT '+sex_zp+' -SATUR_LEVEL '+sex_satur_level+' -MEMORY_PIXSTACK '+sex_mem_pixstack+' -CHECKIMAGE_TYPE '+sex_check_type+' -CHECKIMAGE_NAME '+sex_check_name+' -CATALOG_NAME '+sex_cat_name+' '+sex_im_file
cmd_args=shlex.split(cmd)
print cmd
proc = subprocess.Popen(cmd_args,stdout=subprocess.PIPE, stderr=subprocess.PIPE)
stdout, stderr = proc.communicate()
print stderr
```

```python
sources = root + 'science_reduced_lab_cat.dat'
data = ascii.read(sources, format = "sextractor")  
print data
```

```python
gv=(data["X_IMAGE"] > 100) & (data["X_IMAGE"] < 2000) & (data["MAG_AUTO"] < 99) & (data["FLAGS"] <= 3)

ra = np.array(data["ALPHA_J2000"][gv])
dec = np.array(data["DELTA_J2000"][gv])
flux = np.array(data["FLUX_AUTO"][gv])
mag = np.array(data["MAG_AUTO"][gv])
ellip = np.array(data["ELLIPTICITY"][gv])
theta = np.array(data["THETA_IMAGE"][gv])
x_image = np.array(data["X_IMAGE"][gv])
y_image = np.array(data["Y_IMAGE"][gv])
kron_radius = np.array(data["KRON_RADIUS"][gv])
a_image = np.array(data["A_IMAGE"][gv])
b_image = np.array(data["B_IMAGE"][gv])
```

```python
sorted = np.argsort(mag)
sorted, ra[sorted],dec,mag[sorted],ellip[sorted],theta[sorted],x_image[sorted],y_image[sorted], kron_radius[sorted], a_image[sorted], b_image[sorted]
```

```python
im_data = fits.getdata(root + 'science_reduced_lab_wcs.fits')
im_h=fits.getheader(root + 'science_reduced_lab_wcs.fits')

matplotlib.rcParams['figure.figsize'] = (40.0, 32.0)
fig, ax = plt.subplots()
im = ax.imshow(np.log(im_data),cmap=plt.get_cmap('hot'), interpolation='nearest', origin='lower',
               vmin=np.log(2100), vmax=np.log(2500))
for i in range(0,200):
           ellipse = Ellipse(xy=(x_image[sorted[i]], y_image[sorted[i]]), width=kron_radius[sorted[i]]*a_image[sorted[i]], height=kron_radius[sorted[i]]*b_image[sorted[i]], angle = theta[sorted[i]], edgecolor='c', fc='None', lw=2)
           ax.add_patch(ellipse)
plt.show()
matplotlib.rcParams.update(inline_rc)
```

### We will use the 200 brightest non-saturated sources detected by SExtractor and crossmath their coordinates with the SDSS catalog

```python
gal_sdss_data = np.zeros((200,3))
for i in range(0,200):
     query = "SELECT s.PSFMAG_R, s.RA, s.DEC FROM (Star as s JOIN dbo.fGetNearbyObjEq(" +str(ra[sorted[i]]) + "," + str(dec[sorted[i]]) + ",0.1) AS GN ON s.objID = GN.objID)" 
     result = sqlcl.query(query).readlines()
     if len(result) > 3:
            print 'oops! More than 1 candidate found'
     if len(result) ==2:
            print 'No targets found'
            gal_sdss_data[i]=([0,0,0])
            continue
     gal_sdss_data[i]=(map(float,result[2][:-1].split(',')))
     print 'Done with star',i,' of 200'
                                      
                                                                                                
```

```python
plt.figure(figsize=(12,12))
plt.plot(gal_sdss_data[:,0], mag[sorted[0:200]], 'o')
plt.ylabel('MAG_AUTO',fontsize=20)
plt.xlabel('MAG_SDSS',fontsize=20)
plt.tick_params(axis='x', labelsize=15)
plt.tick_params(axis='y', labelsize=15)
plt.xlim(13,21)
plt.ylim(11,19)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.plot(gal_sdss_data[:,0], gal_sdss_data[:,0]+2.5*np.log10(flux[sorted[0:200]]/600.), 'o')
plt.ylabel('MAG_SDSS - MAG_INST',fontsize=20)
plt.xlabel('MAG_SDSS',fontsize=20)
plt.tick_params(axis='x', labelsize=15)
plt.tick_params(axis='y', labelsize=15)
plt.xlim(12,21)
plt.ylim(24.8,26.0)
plt.axes().set_aspect('auto')
plt.show()
```

Now, just a fit a line and you have the sextractor calibrated magnitudes.

```python
zp_median=np.median(gal_sdss_data[:,0]+2.5*np.log10(flux[sorted[0:200]]/600.))
zp_biweight=funcs.biweight_location(gal_sdss_data[:,0]+2.5*np.log10(flux[sorted[0:200]]/600.))
print 'ZP - Median ', zp_median
print 'ZP - Biweight location ', zp_biweight
```

```python
x=np.linspace(10,30,10)
plt.figure(figsize=(12,8))
plt.plot(x,np.tile(zp_median,10),'r--', label='Median')
plt.plot(x,np.tile(zp_biweight,10),'r-', label='Biweight')
plt.plot(gal_sdss_data[:,0], gal_sdss_data[:,0]+2.5*np.log10(flux[sorted[0:200]]/600.), 'o')
plt.ylabel('MAG_SDSS - MAG_INST',fontsize=20)
plt.xlabel('MAG_SDSS',fontsize=20)
plt.tick_params(axis='x', labelsize=15)
plt.tick_params(axis='y', labelsize=15)
plt.xlim(12,21)
plt.ylim(24.8,26.0)
plt.legend(loc='upper left',prop={'size':15})
plt.show()
```
